```python
driver = webdriver.Chrome(chromedriver)
actor_d = {}
for i, actor in enumerate(actors.loc[481:, 'Actors']):
    driver.get('https://www.imdb.com/')
    ipt = driver.find_element_by_id('navbar-query')
    ipt.send_keys(actor)
    ipt.send_keys(Keys.RETURN)
    try:
        name = driver.find_element_by_link_text(actor)
        name.click()
        url = driver.current_url
        actor_d[actor] = url
    except:
        continue
    if (i + 1) % 20 == 0:
        print(actor)
    if (i + 1) % 100 == 0:
        urls = pd.DataFrame.from_dict(actor_d, orient='index').reset_index()
        urls.to_csv('actors_urls.csv', mode='a', index=False, header=False)
        actor_d = {}
```
