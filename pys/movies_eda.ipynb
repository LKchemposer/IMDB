```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, LassoCV, Ridge, RidgeCV
from sklearn.cross_validation import cross_val_score
from sklearn.cross_validation import KFold
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
movies = pd.read_csv('movies_cleaned.csv')
```

```python
movies['ReleaseMonth'] = movies['ReleaseMonth'].astype(int)
movies['ReleaseYear'] = movies['ReleaseYear'].astype(int)
```

```python
movies.head()
```

```python
movies.corr()
```

```python
cols = ['IMDBRating', 'IMDBRatingCount', 'RunTime', 'UserCount', 'CriticCount', 'Metascore']
sns.pairplot(movies[cols]);
```

```python
cols = ['IMDBRating', 'Budget', 'Opening', 'GrossUSA', 'GrossIntl', 'Comps']
sns.pairplot(movies[cols]);
```

```python
cols = ['IMDBRating', 'Awards', 'Noms', 'Wins', 'ActAwards', 'ActNoms', 'ActWins', 'ActCreds']
sns.pairplot(movies[cols]);
```

```python
y = movies['IMDBRating']
X = movies.drop('IMDBRating', axis=1)
X[['logUserCount', 'logBudget', 'logOpening', 'logGrossUSA', 'logGrossIntl']] = X[['UserCount', 'Budget', 'Opening', 'GrossUSA', 'GrossIntl']].apply(np.log)
X['logIMDB'] = np.log(X['IMDBRatingCount'])
X['logMetascore'] = np.log(X['Metascore'])
```

```python
select = 'logMetascore'
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
lr = LinearRegression()
lr.fit(X_train[[select]], y_train)
lr.coef_, lr.intercept_
```

```python
plt.scatter(X_train[select], y_train, alpha=0.2);
xs = np.linspace(np.min(X_train[select]), np.max(X_train[select]))
ys = lr.intercept_ + lr.coef_[0] * xs
plt.plot(xs, ys);
```

```python
plt.scatter(X_test[select], y_test, alpha=0.2);
xs = np.linspace(np.min(X_train[select]), np.max(X_train[select]))
ys = lr.intercept_ + lr.coef_[0] * xs
plt.plot(xs, ys);
```

```python
y_pred = lr.predict(X_test[[select]])
plt.scatter(y_pred, y_test, alpha=0.2);
xs = ys = np.linspace(np.min(y_pred), np.max(y_pred)) 
plt.plot(xs, ys);
```

```python
plt.scatter(y_pred, y_test - y_pred, alpha=0.2);
plt.plot(np.linspace(0, 10), np.linspace(0, 0));
```

```python
lr.score(X_test[[select]], y_test)
```

```python
#with Metascore
cols = ['logIMDB', 'RunTime', 'logOpening', 'logUserCount', 'logMetascore', 'CriticCount', 'logGrossUSA', 'logBudget']
X2 = (X[cols].join(movies['Genres'].str.get_dummies(', '), how='inner'))
```

```python
X_tv, X_test, y_tv, y_test = train_test_split(X2, y, test_size=0.2)
X_train, X_val, y_train, y_val = train_test_split(X_tv, y_tv, test_size=0.25)
```

```python
std = StandardScaler()
std.fit(X_train);
X_train = std.transform(X_train)
X_val = std.transform(X_val);
X_test = std.transform(X_test);
```

```python
alphas = 10 ** np.linspace(-10, 10, 21)
lcv = LassoCV(alphas = alphas, cv=10)
lcv.fit(X_train, y_train)
y_pred = lcv.predict(X_val)
lcv.score(X_val, y_val), lcv.alpha_, np.sqrt(mean_squared_error(y_val, y_pred))
```

```python
rcv = RidgeCV(alphas=alphas, cv=10)
rcv.fit(X_train, y_train)
y_pred = rcv.predict(X_val)
rcv.score(X_val, y_val), rcv.alpha_, np.sqrt(mean_squared_error(y_val, y_pred))
```

```python
plt.scatter(y_pred, y_val - y_pred, alpha=0.2);
plt.plot(np.linspace(0, 10), np.linspace(0, 0));
```

```python
y_pred = lcv.predict(X_test)
lcv.score(X_test, y_test), np.sqrt(mean_squared_error(y_test, y_pred))
```
